```python
import numpy as np

# Step 1: Generate a random dataset of 100 points
np.random.seed(42)  # For reproducibility
dataset = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Standard Deviation=10

# Step 2: Calculate mean and standard deviation
mean_value = np.mean(dataset)
std_deviation = np.std(dataset)

# Step 3: Standardize the values and express as percentages
standardized_percentages = [((x - mean_value) / std_deviation) * 100 for x in dataset]

# Print the first 10 standardized values as percentages
print("First 10 Standardized Values as Percentages:")
print(standardized_percentages[:10])
```

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate a random dataset of 100 points
np.random.seed(42)  # For reproducibility
dataset = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Standard Deviation=10

# Step 2: Calculate mean and standard deviation
mean_value = np.mean(dataset)
std_deviation = np.std(dataset)

# Calculate the z-scores (standardized values)
z_scores = (dataset - mean_value) / std_deviation

# Calculate the percentage of values within each standard deviation
percent_within_std = [stats.norm.cdf(i) * 100 for i in z_scores]

# Print the percentage of values within each standard deviation
for i, percent in enumerate(percent_within_std):
    print(f"Percentage within {i + 1} standard deviation: {percent:.2f}%")
```
